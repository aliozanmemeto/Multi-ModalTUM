# Notebook 2: Text to speech (TTS)

In this notebook, we will generate a brief sound clip for our video. We will use [Kokoro TTS](https://github.com/hexgrad/kokoro), a very lightweight (82M) model that punches far above its weight. You can find details about the model [here](https://huggingface.co/hexgrad/Kokoro-82M#model-facts) if you are interested. The model is also permissively licensed and multi-lingual.

Note that, in a full video generation pipeline, the TTS model would either be integrated with the video generation model or at least conditioned on the video and not just on a prompt, but there are unfortunately no current free-and-open-source video models that can do this, and they would be far too computationally expensive to run on Google Colab anyway.

Kokoro uses a pipeline similar to Diffusers. The usage is straightforward. You can change the voices and languages to create something that is non-english. See the documentation for the various voices [here](https://huggingface.co/hexgrad/Kokoro-82M/blob/main/VOICES.md).

Make sure you save the generated wav file since we will need it for the next notebook. Ideally, don't generate audio that's much longer than the video you generated in Notebook 1.

```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf


pipeline = KPipeline(lang_code='a')
text = '''
Wow! This cat smells like bacon!
'''
generator = pipeline(text, voice='af_heart')
for i, (gs, ps, audio) in enumerate(generator):
    print(i, gs, ps)
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{i}.wav', audio, 24000)
```
